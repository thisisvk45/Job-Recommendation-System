```python
import nltk
from nltk.tokenize import word_tokenize
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
from string import punctuation

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    word_sent = word_tokenize(text.lower().replace("\n", ""))
    stop_words = set(stopwords.words('english') + list(punctuation))
    lemmatizer = WordNetLemmatizer()
    processed_text = [lemmatizer.lemmatize(word) for word in word_sent if word not in stop_words]
    return ' '.join(processed_text)

df['Processed Description'] = df['Description'].apply(preprocess_text)

print(df[['Description', 'Processed Description']].head())
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf_vectorizer = TfidfVectorizer()
tfidf_matrix = tfidf_vectorizer.fit_transform(df['Processed Description'])
print(tfidf_matrix.shape)
```

```python
from sklearn.feature_extraction.text import CountVectorizer

count_vectorizer = CountVectorizer()

count_matrix = count_vectorizer.fit_transform(df['Processed Description'])

cv_count = count_vectorizer.transform([processed_cv_text])

cosine_similarities_count = cosine_similarity(cv_count, count_matrix).flatten()

top_indices_count = cosine_similarities_count.argsort()[-5:][::-1]

recommended_jobs_count = df.iloc[top_indices_count][['Job Opening Title', 'Company Name', 'Job Location', 'Estimated Salary (INR)', 'Description']]
```

```python
from sklearn.neighbors import NearestNeighbors
from sklearn.feature_extraction.text import TfidfVectorizer

knn_tfidf_vectorizer = TfidfVectorizer(stop_words='english')
knn_matrix = knn_tfidf_vectorizer.fit_transform(df['Processed Description'])

knn_model = NearestNeighbors(n_neighbors=5, algorithm='auto').fit(knn_matrix)
distances, indices = knn_model.kneighbors(knn_tfidf_vectorizer.transform([processed_cv_text]))

recommended_jobs_knn = df.iloc[indices[0]][['Job Opening Title', 'Company Name', 'Job Location', 'Estimated Salary (INR)', 'Description']]
```

```python
import numpy as np
from sklearn.preprocessing import MinMaxScaler

df_tfidf = pd.DataFrame(cosine_similarities, columns=['TF-IDF'])
df_count = pd.DataFrame(cosine_similarities_count, columns=['CountVectorizer'])
df_knn = pd.DataFrame(np.zeros(len(df)), columns=['KNN'])
df_knn.iloc[indices[0], 0] = 1 - distances[0]

combined_df = pd.concat([df_tfidf, df_count, df_knn], axis=1)

scaler = MinMaxScaler()
combined_df[['TF-IDF', 'CountVectorizer', 'KNN']] = scaler.fit_transform(combined_df[['TF-IDF', 'CountVectorizer', 'KNN']])

weights = {'TF-IDF': 0.3, 'CountVectorizer': 0.3, 'KNN': 0.4}
combined_df['FinalScore'] = (combined_df['TF-IDF'] * weights['TF-IDF'] +
                             combined_df['CountVectorizer'] * weights['CountVectorizer'] +
                             combined_df['KNN'] * weights['KNN'])

top_indices_final = combined_df['FinalScore'].argsort()[-5:][::-1]

final_recommendations = df.iloc[top_indices_final][['Job Opening Title', 'Company Name', 'Job Location', 'Estimated Salary (INR)', 'Description']]
```
